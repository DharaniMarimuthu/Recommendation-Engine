# ITEM AFFINITY

```python
import nltk
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
stop_words = set(stopwords.words('english'))
```

```python
import string
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
import math
```

```python
anime.head()
```

```python
affinity_dataset=anime[['anime_id','title','synopsis','score']]
```

```python
affinity_dataset.head()
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(affinity_dataset,test_size=0.5)
```

```python
def word_split(dataset):
    dataset['synopsis']=dataset['synopsis'].astype(str).str.lower().str.split()
    dataset['synopsis']=dataset['synopsis'].apply(lambda x: [item for item in x if item not in stop_words])
    return dataset
```

```python
def count_words(dataset):
    count=pd.Series([float])
    count=affinity_dataset['synopsis'].str.len()
    dataset.insert(3,'count',count)
    words_count=dataset['synopsis'].apply(pd.value_counts)
    print("words_count:",words_count.shape)
    return words_count
```

```python
def TF_IDF(dataset,words_count):
    TF_matrix=words_count.div(dataset['count'],axis=0)
    print("TF_matrix:",TF_matrix.shape)
    IDF_matrix = TF_matrix.notnull().sum()
    IDF=np.log(TF_matrix.shape[0]/IDF_matrix)
    IDF=IDF.round(2)
    print("IDF:",IDF.shape)
    F=TF_matrix*IDF
    F=F.fillna(0).round(2)
    print("F:",F.shape)
    return F
```

```python
def cos_sim(F):
    sim=cosine_similarity(F)
    similarity=pd.Series([float])
    similarity=sim.mean(axis=0)
    return similarity
```

```python
def prediction(data):
    rating=review_dataset[['anime_id','score']]
    rating=rating.groupby('anime_id').mean()
    dataset=pd.merge(data[['anime_id','title','synopsis','score']],rating,on='anime_id')
    dataset=word_split(dataset)
    words_count=count_words(dataset)
    F=TF_IDF(dataset,words_count)
    similarity=cos_sim(F)
    pred=pd.Series([int])
    pred=(dataset['score_y']*similarity)/similarity
    dataset.insert(3,'pred',pred)
    dataset.sort_values(by=['pred'],inplace=True,ascending=False)
    recommendation=dataset[['anime_id','title','score_x','pred']]
    return recommendation
```

```python
result_train=prediction(train)
print(result_train.head())
```

```python
rmse=math.sqrt(((result_train['pred']-result_train['score_x'])**2).mean())
```

```python
rmse
```

```python
np.round(rmse,decimals=3)
```

```python
test_set=test.append(train)
result_test=prediction(test_set)
print(result_test.head())
```

```python
rmse_test=math.sqrt(((result_test['pred']-result_test['score_x'])**2).mean())
```

# matrix factorization

```python
from sklearn.model_selection import train_test_split
from scipy.sparse.linalg import svds
from sklearn.metrics.pairwise import cosine_similarity
```

```python
mf_dataset=pd.merge(review_dataset[['user_id','anime_id','score']],anime[['anime_id','title']],on='anime_id')
```

```python
training,test=train_test_split(mf_dataset,test_size=0.5)
```

```python
def MF(train):
    R_df=train.pivot_table('score','user_id','title',fill_value=0)
    R = R_df.to_numpy()

    user_ratings_mean = np.mean(R, axis = 1)
    R_demeaned = R - user_ratings_mean.reshape(-1, 1)

    P, sigma, Q = svds(R_demeaned, k = 3)

    sigma = np.diag(sigma)

    all_user_predicted_ratings = np.dot(np.dot(P, sigma), Q) + user_ratings_mean.reshape(-1, 1)
    pred_df = pd.DataFrame(all_user_predicted_ratings, columns = R_df.columns)

    preds_df=pred_df.to_numpy()

    alpha=0.1
    beta=0.1

    U=np.ones([P.shape[0],P.shape[1]],dtype=int)/np.sqrt(P.shape[1])

    A=np.ones([Q.shape[0],Q.shape[1]],dtype=int)/np.sqrt(Q.shape[0])

    for j in range(10):
        for i in range(len(A)):
            Ai=np.transpose(U)@U+alpha*np.identity(U.shape[1])
            Vi=np.transpose(U)@preds_df.T[i]
            A.T[i]=np.linalg.pinv(Ai) @ Vi

        for i in range(len(U)):
            Ai=A@np.transpose(A)+beta*np.identity(A.shape[0])
            Vi=A@preds_df[i]
            U[i]=np.linalg.pinv(Ai) @ Vi

    sample=np.dot(np.dot(U, sigma), A) + user_ratings_mean.reshape(-1, 1)
    return sample,pred_df,R
```

```python
def hybrid(train):
    sample,pred_df,R=MF(train)
    Ai=cosine_similarity(sample).mean(axis=0)

    binary=pd.Series(int)

    binary=(train['score']/8.5).astype(int)

    train.insert(4,'binary_rat',binary)

    one=(len(train[train.binary_rat==1]))/len(train)

    bin_rat=train.pivot_table('binary_rat','user_id','title',fill_value=0)

    Ai_one=((bin_rat)/len(train[train.binary_rat==1]))
    Ai_one=Ai_one*one
    pred=Ai_one.divide(Ai,axis=0)
    pred_df.replace(to_replace='score',value='pred')
    
    R=R.flatten()
    preds=pred_df.to_numpy()
    preds=preds.flatten()
    r=R-preds

    import math
    print(np.sqrt(sum(r**2))/len(r))

    return pred_df
```

```python
def recommend(train,UserID):
    pred_df=hybrid(train)
    recommendations=pred_df.iloc[UserID].sort_values(ascending=False).head(5)
    print(recommendations)
```

```python
print('training set..')
result_train=recommend(training,1)
```
